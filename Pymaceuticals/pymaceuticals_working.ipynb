## Observations and Insights

The following conclusions can be drawn from the drug regimen and animal study data provided by Pymaceuticals:
•	Of the four drug regimens (Capomulin, Ramicane, Infubinol, and Ceftamin) Capomulin and Ramicane seem to be more 
    effective for treating the mice with SCC tumor growth based on the final tumor volume. 

•	The mice with the tumor growth, that are included in the study seem to be pretty equally distributed in terms of 
    gender. It would be interesting to see if gender impacts the efficacy of the drug regimen.

•	As time progresses, the tumor volume in mice seems to shrink down consistently with the use of Capomulin regimen. 

•	A high R-squared of 0.84 suggests a strong correlation between the mouse weight and the average tumor volume 
    indicating the weight of the mice being treated also has an impact on the efficacy of the drug.

•	More data such as tumor development beyond the 45-day period, starting tumor volumes in the mice being treated 
    would help in determining the efficacy of the drug regimens.

## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
mouse_study_data = pd.merge(left=mouse_metadata,right=study_results,how="outer",on=['Mouse ID'])
display(mouse_study_data.head(10))
#print(mouse_study_data)
#print(mouse_study_data.isnull().any().any())
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
sum_stats = mouse_study_data.groupby('Drug Regimen').aggregate({"Tumor Volume (mm3)":['mean', 'median', 'var', 'std','sem']}) 
sum_stats = sum_stats.rename(columns={
    "mean": "Mean", "median": "Median", "var": "Variance","std": "Standard Deviation","sem": "Standard Error"
})
#print(sum_stats)
display(sum_stats)
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers.
pop_regimen_data = mouse_study_data[mouse_study_data['Drug Regimen'].str.contains('Capomulin', case=False) | mouse_study_data['Drug Regimen'].str.contains('Ramicane', case=False) | mouse_study_data['Drug Regimen'].str.contains('Infubinol', case=False) | mouse_study_data['Drug Regimen'].str.contains('Ceftamin', case=False)]
tumor_vol_data = pop_regimen_data.groupby(['Drug Regimen','Mouse ID']).aggregate({'Tumor Volume (mm3)': 'last'})
tumor_series = tumor_vol_data.iloc[:,0]
final_tumor_vol = tumor_series.to_list()
new_df = pop_regimen_data.groupby(['Drug Regimen','Mouse ID']).size().reset_index(name='count')
new_df['Final Tumor Volume'] = final_tumor_vol
new_df = new_df.drop(['count'], axis=1)
#print(new_df)
display(new_df)

cap_df = new_df[new_df['Drug Regimen'].str.contains('Capomulin')]
#print(cap_df)
cap_quartiles = cap_df['Final Tumor Volume'].quantile([.25,.5,.75])
cap_lowerq = cap_quartiles[0.25]
cap_upperq = cap_quartiles[0.75]
cap_iqr = cap_upperq-cap_lowerq
print(f"The lower quartile of Tumor Volume (mm3) for Capomulin is: {cap_lowerq}")
print(f"The upper quartile of Tumor Volume (mm3) for Capomulin is: {cap_upperq}")
print(f"The interquartile range of Tumor Volume (mm3) for Capomulin is: {cap_iqr}")
print(f"The the median of Tumor Volume (mm3) for Capomulin is: {cap_quartiles[0.5]} ")
cap_lower_bound = cap_lowerq - (1.5*cap_iqr)
cap_upper_bound = cap_upperq + (1.5*cap_iqr)
print(f"Values below {cap_lower_bound} could be outliers for Capomulin.")
print(f"Values above {cap_upper_bound} could be outliers for Capomulin.")
print(f"-----------------------------------------------------------------------------")

ram_df = new_df[new_df['Drug Regimen'].str.contains('Ramicane')]
#print(ram_df)
ram_quartiles = ram_df['Final Tumor Volume'].quantile([.25,.5,.75])
ram_lowerq = ram_quartiles[0.25]
ram_upperq = ram_quartiles[0.75]
ram_iqr = ram_upperq-ram_lowerq
print(f"The lower quartile of Tumor Volume (mm3) for Ramicane is: {ram_lowerq}")
print(f"The upper quartile of Tumor Volume (mm3) for Ramicane is: {ram_upperq}")
print(f"The interquartile range of Tumor Volume (mm3) for Ramicane is: {ram_iqr}")
print(f"The the median of Tumor Volume (mm3) for Ramicane is: {ram_quartiles[0.5]} ")
ram_lower_bound = ram_lowerq - (1.5*ram_iqr)
ram_upper_bound = ram_upperq + (1.5*ram_iqr)
print(f"Values below {ram_lower_bound} could be outliers for Ramicane.")
print(f"Values above {ram_upper_bound} could be outliers for Ramicane.")
print(f"-----------------------------------------------------------------------------")

inf_df = new_df[new_df['Drug Regimen'].str.contains('Infubinol')]
#print(inf_df)
inf_quartiles = inf_df['Final Tumor Volume'].quantile([.25,.5,.75])
inf_lowerq = inf_quartiles[0.25]
inf_upperq = inf_quartiles[0.75]
inf_iqr = inf_upperq-inf_lowerq
print(f"The lower quartile of Tumor Volume (mm3) for Infubinol is: {inf_lowerq}")
print(f"The upper quartile of Tumor Volume (mm3) for Infubinolis: {inf_upperq}")
print(f"The interquartile range of Tumor Volume (mm3) for Infubinol is: {inf_iqr}")
print(f"The the median of Tumor Volume (mm3) for Infubinol is: {inf_quartiles[0.5]} ")
inf_lower_bound = inf_lowerq - (1.5*inf_iqr)
inf_upper_bound = inf_upperq + (1.5*inf_iqr)
print(f"Values below {inf_lower_bound} could be outliers for Infubinol.")
print(f"Values above {inf_upper_bound} could be outliers for Infubinol.")
print(f"-----------------------------------------------------------------------------")

cef_df = new_df[new_df['Drug Regimen'].str.contains('Ceftamin')]
#print(cef_df)
cef_quartiles = cef_df['Final Tumor Volume'].quantile([.25,.5,.75])
cef_lowerq = cef_quartiles[0.25]
cef_upperq = cef_quartiles[0.75]
cef_iqr = cef_upperq-cef_lowerq
print(f"The lower quartile of Tumor Volume (mm3) for Ceftamin is: {cef_lowerq}")
print(f"The upper quartile of Tumor Volume (mm3) for Ceftamin is: {cef_upperq}")
print(f"The interquartile range of Tumor Volume (mm3) for Ceftamin is: {cef_iqr}")
print(f"The the median of Tumor Volume (mm3) for Ceftamin is: {cef_quartiles[0.5]} ")
cef_lower_bound = cef_lowerq - (1.5*cef_iqr)
cef_upper_bound = cef_upperq + (1.5*cef_iqr)
print(f"Values below {cef_lower_bound} could be outliers for Ceftamin.")
print(f"Values above {cef_upper_bound} could be outliers for Ceftamin.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
data_to_plot = [cap_df['Final Tumor Volume'], ram_df['Final Tumor Volume'], inf_df['Final Tumor Volume'], cef_df['Final Tumor Volume']]             
fig = plt.figure(1, figsize=(9, 6))

# Create axes instance
ax = fig.add_subplot(111)
ax.set_title('Final tumor volume of each mouse across four drug regimens',fontsize=14)
ax.set_xlabel('Drug Regimen')
ax.set_ylabel('Final Tumor Volume (mm3)')

# Create the boxplot
ax.set_xticklabels(['Capomulin', 'Ramicane', 'Infubinol','Ceftamin'])
bp = ax.boxplot(data_to_plot, showfliers=True)

for flier in bp['fliers']:
    flier.set(marker='o', markerfacecolor='b', markersize=12, markeredgecolor='g', alpha=0.5)

plt.show()
fig.savefig('Tumor_BoxPlot.png', bbox_inches='tight')
```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
cap_regimen_data = mouse_study_data[mouse_study_data['Drug Regimen'].str.contains('Capomulin')]
time_tumor_cap_data = cap_regimen_data.groupby(['Mouse ID','Timepoint','Tumor Volume (mm3)']).size().reset_index(name='count')
mouse_time_tumor_cap_data = time_tumor_cap_data[time_tumor_cap_data['Mouse ID'].str.contains('b128')]
mouse_time_tumor_cap_data.plot(x ='Timepoint', y='Tumor Volume (mm3)', kind='line', marker='o', figsize=(8, 6), linewidth=2.5)
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Time point vs Tumor volume for a mouse treated with Capomulin", fontsize=14)
plt.grid(True)
plt.savefig("LinePlot.png")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
mouse_tumor_cap_data = cap_regimen_data.groupby(['Mouse ID','Weight (g)'], as_index=False)['Tumor Volume (mm3)'].mean()
#print(mouse_tumor_cap_data)
mouse_wt = mouse_tumor_cap_data['Weight (g)']
avg_tumor_vol = mouse_tumor_cap_data['Tumor Volume (mm3)']
plt.figure(figsize=(8, 6))
plt.scatter(mouse_wt,avg_tumor_vol, marker="o", c="red", edgecolors= "black", s=40,label= 'Tumor Volume (mm3)')
plt.legend(loc="best")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse weight vs Average tumor volume for Capomulin", fontsize=14)
plt.grid(True)
plt.savefig("ScatterPlot.png")
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(mouse_wt,avg_tumor_vol)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(mouse_wt, avg_tumor_vol)
regress_values = mouse_wt * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.figure(figsize=(8, 6))
plt.scatter(mouse_wt,avg_tumor_vol,s=40, c='red',edgecolors='black',label= 'Tumor Volume (mm3)')
plt.legend(loc="best")
plt.plot(mouse_wt,regress_values,"r-",color='blue')
plt.annotate(line_eq,(17,37),fontsize=15,color="black")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse weight vs Average tumor volume for Capomulin with Linear Regression", fontsize=14)
plt.grid(True)
plt.savefig("Scatter_LinReg_Plot.png")
plt.show()
```
